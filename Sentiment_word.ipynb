Traffic Data Analysis

Based on recent reports of cases to the SFPD, we are running a wordcloud to see what words are most common in the reports.

```python
#%pip install python-docx wordcloud matplotlib

from docx import Document
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Function to extract text from a .docx file
def extract_text_from_docx(docx_path):
    doc = Document(docx_path)
    full_text = []
    for para in doc.paragraphs:
        full_text.append(para.text)
    return '\n'.join(full_text)

# Extract text from the .docx file
docx_path = 'wordcloud txt.docx'
text = extract_text_from_docx(docx_path)

# Create and generate a word cloud image:
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the generated image:
plt.figure(figsize=(15, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Drilling down to top 15 words for the wordcloud

```python
#%pip install python-docx wordcloud matplotlib

from docx import Document
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Function to extract text from a .docx file
def extract_text_from_docx(docx_path):
    doc = Document(docx_path)
    full_text = []
    for para in doc.paragraphs:
        full_text.append(para.text)
    return '\n'.join(full_text)

# Extract text from the .docx file
docx_path = 'wordcloud txt.docx'  # Replace with your .docx file path
text = extract_text_from_docx(docx_path)

# Create and generate a word cloud image:
wordcloud = WordCloud(width=800, height=400, background_color='white')
wordcloud.generate(text)

# Calculate word frequencies using the generate_from_frequencies method
word_frequencies = wordcloud.process_text(text)
sorted_words = sorted(word_frequencies.items(), key=lambda x: x[1], reverse=True)

# Display the top 10 words by frequency
top_words = sorted_words[:15]
for word, freq in top_words:
    print(f"Word: {word}, Frequency: {freq}")

# Generate the word cloud with frequencies (this could be commented out if not needed)
wordcloud = wordcloud.generate_from_frequencies(dict(top_words))

# Display the generated image
plt.figure(figsize=(15, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
